Sigmoid Perceptron Class

We will be using stocastic GD optimization algo. to update the weights and bias after training with each data point

```python
class SigmoidPerceptron():

# initiate the parameters
  def __init__(self):                                              #self: when calling the function of a class within other functions
                                                                    # input_size: shape of the data ; # No. weights = No. input features
    self.weights = np.random.randn(input_size)                      # initiating a weight array so as to have same number of values as the no. columns
    self.bias = np.random.randn(1)                                  # (1) as bias is single scalar value

# Give the sigmoid formula
  def sigmoid(self, z):                                             # 'z' is the weighted sum
    return 1/(1 + np.exp(-z))


# Calculates weighted sum and add the bias then call the sigmoid
  def predict(self, input):

    weighted_sum = np.dot(inputs, self.weights) + self.bias        # dot product of vectors is always scalar
    return self.sigmoid(weighted_sum)

# Update the weight using Stocastic GD                             # Optimizing and updating the weights and bias
  def fit(self, inputs, targets, learning_rate, num_epochs):       # targets -> labels ; learning_rate: how much weight to give to weight & bias at each iterations

    number_examples = input.shape[0]

    for epoch in range(num_epochs):                                # In standard GD, GD takes place here

      for i in range(num_examples):                                # This is the addition for S.GD, take 1 data point and predict then compare to true label
                                                                   # GD takes place here
        input_vector = inputs[i]
        prediction = self.predict(input_vector)
        error = targets[i] - prediction

# Updating Weights (dw)

        gradient_weights = error * prediction * (1-prediction) * input_vector
        self.weights += learning_rate * gradient_weights

# Updating Bias  (db)

        gradient_bias = error * prediction * (1-prediction)
        self.bias += learning_rate * gradient_bias



# Accuracy of the model                                # Accuracy = No. correct predictions / total no. data points
  def evaluate(self, inputs, targets):

    correct = 0

    for input_vector, target in zip(inputs, targets):  # zip: iterating for loop through 2 lists (input_vector -> List1 ; target -> List2)
      prediction = self.predict(input_vector)

      if prediction >= 0.5:
        predicted_class = 1
      else:
        predicted_class = 0

      if predicted_class == target:
        correct += 1

    accuracy = correct / len(inputs)
    return accuracy
```
